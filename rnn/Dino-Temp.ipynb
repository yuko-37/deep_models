```python
import pickle
import numpy as np
```

```python
with open('dino-parameters', 'rb') as file:
    parameters = pickle.load(file)
```

```python
print(parameters.keys())
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

def sample(parameters):
    Waa, Wax, Wya, by, b = parameters['Waa'], parameters['Wax'], parameters['Wya'], parameters['by'], parameters['b']
    vocab_size = by.shape[0]
    n_a = Waa.shape[1]

    x = np.zeros((vocab_size, 1))
    a_prev = np.zeros((n_a, 1))
    indices = []
    idx = -1 
    
    counter = 0
    newline_character = 0
    
    while (idx != newline_character and counter != 50):
        a = np.tanh(Wax @ x + Waa @ a_prev + b)
        z = Wya @ a + by
        y = softmax(z)
        
        idx = np.random.choice(range(len(y)), p=y.ravel())
        if (idx != newline_character):
            indices.append(idx)
        x = np.zeros((vocab_size, 1))
        x[idx] = 1
        a_prev = a
        counter +=1

    
    return indices
```

```python
indices = sample(parameters)
```

```python
chars = ['\n', 'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
ix_to_char = {i:ch for i, ch in enumerate(chars)}
print(ix_to_char)
```

```python
def get_sample(parameters, ix_to_char):
    sampled_ix = sample(parameters)
    sampled_chars = [ix_to_char[i] for i in sampled_ix]
    generated_name = "".join(sampled_chars)

    return generated_name
```

```python
def get_sample(parameters, ix_to_char):
    sample_ix = sample(parameters)
    txt = ''.join(ix_to_char[ix] for ix in sample_ix)
    txt = txt[0].upper() + txt[1:]
    return txt
```

```python
for i in range(5):
    print(get_sample(parameters, ix_to_char))
```
